```python
# Load the Iris dataset
iris = load_iris()

# Print the keys of the Iris dataset
print("Keys in Iris dataset:", iris.keys())

# Print the number of rows and columns in the data
n_samples, n_features = iris.data.shape
print(f"Number of rows (samples): {n_samples}")
print(f"Number of columns (features): {n_features}")

# Print the feature names
print("Feature names:")
for feature_name in iris.feature_names:
    print("-", feature_name)

# Print the description of the Iris dataset
print("Iris dataset description:")
print(iris.DESCR)
```

```python
# Perform one-hot encoding on the 'species' column
df2_encoded = pd.get_dummies(df2, columns=['Species'])

# Display the DataFrame with one-hot encoding
print(df2_encoded)
```

```python
# Drop the "id" column for each species
df2_setosa = df2[df2['Species'] == 'setosa'].drop(columns=['Id'])
df2_versicolor = df2[df2['Species'] == 'versicolor'].drop(columns=['Id'])
df2_virginica = df2[df2['Species'] == 'virginica'].drop(columns=['Id'])

# Display or save the modified DataFrames as needed
print("Iris Data Without 'Id' column for Setosa:")
print(df2_setosa)

print("\nIris Data Without 'Id' column for Versicolor:")
print(df2_versicolor)

print("\nIris Data Without 'Id' column for Virginica:")
print(df2_virginica)
```
